```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#Open Wages data
dfw = pd.read_csv("data/manufacturing-wages-mexico-total.csv",skiprows=4)
```

```python
#Show data
dfw
```

```python
#drop final rows
dfw = dfw.drop([174,175,176])
```

```python
dfw
```

```python
#untidy data
dfw_melted = pd.melt(dfw, id_vars=["Periodos"], var_name ="State", value_name ="Manufacturing wages").copy()
dfw_melted
```

```python
rowsw = dfw_melted.shape[0]
rowsw
```

```python
#clean data (parentesis)
for j in range(rowsw):
    dfw_melted["State"][j]= re.sub(r'\([^)]*\)', '', dfw_melted["State"][j])
```

```python
 dfw_melted["State"][1]
```

```python
#clean data (other text)
dfw_clean= dfw_melted.replace("Manufacturas > Industria manufacturera, maquiladora y de servicios de exportaci�n  > Por entidad federativa > Establecimientos manufactureros > Remuneraciones reales por persona ocupada seg�n entidades federativas > Salarios y sueldos pagados al personal > Total > ","",regex=True)
dfw_clean= dfw_clean.replace("/a /f1","",regex=True)
dfw_clean= dfw_clean.replace(">",",",regex=True)
dfw_clean
```

```python
#split columns
dfw_clean["Month"] = dfw_clean["Periodos"].str.split('/', expand=True)[1]
dfw_clean["Year"] = dfw_clean["Periodos"].str.split('/', expand=True)[0]
dfw_clean.drop("Periodos", axis=1)
dfw_clean
```

```python
#arrange columns
dfw_clean = dfw_clean[["Year","Month", "State", "Manufacturing wages"]]
dfw_clean
```

```python
dfw_clean["State"].unique()
```

```python
#clean other characters
dfw_clean = dfw_clean.replace("M�xico","Mexico",regex=True)
dfw_clean = dfw_clean.replace("Le�n","Leon",regex=True)
dfw_clean = dfw_clean.replace("Quer�taro","Queretaro",regex=True)
dfw_clean = dfw_clean.replace("Potos�","Potosi",regex=True)
dfw_clean = dfw_clean.replace("Yucat�n","Yucatan",regex=True)
```

```python
dfw_clean["State"].unique()
```

```python
#remove blank spaces at the begining and the end
for q in range(rowsw):
    dfw_clean["State"][q] = dfw_clean["State"][q].rstrip().lstrip()
```

```python
#remove blank spaces at the end
for d in range(rowsw):
    dfw_clean["State"][d] = dfw_clean["State"][d].rstrip()
```

```python
#flag border states
border = ["Baja California", "Coahuila de Zaragoza", "Chihuahua", "Nuevo Leon", "Tamaulipas", "Sonora"]
f = len(border)
dfw_clean["Border"] = "NA"
for g in range(rowsw):
    p = 0
    for h in range(f):
        if dfw_clean["State"][g] == border[h]:
            p = p +1
    if p > 0:
        dfw_clean["Border"][g] = "Yes"
    else:
        dfw_clean["Border"][g] = "No"
```
