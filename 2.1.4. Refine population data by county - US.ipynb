```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#open manufacturing workers data
df1 = pd.read_csv("data/population-by-count-US-2007-2010-raw.csv", skiprows=3)
df2 = pd.read_csv("data/population-by-count-US-2011-2015-raw.csv", skiprows=3)
df3 = pd.read_csv("data/population-by-count-US-2016-2021-raw.csv", skiprows=3)
```

```python
#show data
df3.tail(20)
```

```python
#drop final rows
df1 = df1.drop([3140,3141,3142,3143,3144,3145,3146,3147,3148,3149,3150,3151,3152,3153,3154,3155,3156,3157,3158])
df2 = df2.drop([3140,3141,3142,3143,3144,3145,3146,3147,3148,3149,3150,3151,3152,3153,3154,3155,3156,3157,3158])
df3 = df3.drop([3140,3141,3142,3143,3144,3145,3146,3147,3148,3149,3150,3151,3152,3153,3154,3155,3156,3157,3158])
df3
```

```python
#untidy data
df1_melted = pd.melt(df1, id_vars=["GeoFips","GeoName"]).copy()
df2_melted = pd.melt(df2, id_vars=["GeoFips","GeoName"]).copy()
df3_melted = pd.melt(df3, id_vars=["GeoFips","GeoName"]).copy()
```

```python
#concat data
df = pd.concat([df1_melted,df2_melted,df3_melted], axis = 0)
```

```python
df
```

```python
df["value"] = pd.to_numeric(df["value"], errors='coerce')
df.info()
```

```python
#split columns
df["County"] = df["GeoName"].str.split(',', expand=True)[0]
df["State"] = df["GeoName"].str.split(',', expand=True)[1]
```

```python
df
```

```python
#arrange columns
df = df.drop(["GeoFips","GeoName"], axis =1)
df.rename(columns = {"variable": "Year","value": "Value" }, inplace = True)
df["Description"] = "Population"
df = df[["Year", "Description", "County","State", "Value"]]

df
```

```python
rows = df.shape[0]
rows
```

```python
df = df.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)
df = df.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)
```

```python
df = df.reset_index(drop=True)
df
```

```python
df["State"].unique()
```

```python
#remove characters
df= df.replace("AK*","AK")
df= df.replace("AZ*","AZ")
df= df.replace("HI*","HI")
df= df.replace("NM*","NM")
df= df.replace("SD*","SD")
df= df.replace("CO*","CO")
df= df.replace("VA*","VA")
df= df.replace("WI*","WI")
df["State"].unique()
```

```python
#flag border states
border = ["California", "Arizona", "New Mexico", "Texas"]
f = len(border)
df["Border State"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df["State"][k] == border[l]:
            p = p +1
    if p > 0:
        df["Border State"][k] = "Yes"
    else:
        df["Border State"][k] = "No"
```

```python
#flag border counties
border_counties = ['San Diego', 'Imperial', 'Yuma', 'Pima', 'Santa Cruz', 'Cochise', 'Hidalgo', 'Luna' , 'Doña Ana','El Paso','Hudspeth' ,'Jeff Davis','Presidio', 'Brewster', 'Terrell' ,'Val Verde', 'Kinney' ,'Maverick', 'Webb' ,'Zapata', 'Starr' ,'Hidalgo', 'Cameron']
f = len(border_counties)
df["Border County"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df["County"][k] == border_counties[l]:
            p = p +1
    if p > 0:
        df["Border County"][k] = "Yes"
    else:
        df["Border County"][k] = "No"
```

```python
#Remove non states
df = df.query("State != 'Staunton + Waynesboro' \
                                           & State != 'Colonial Heights + Petersburg' \
                                           & State != 'Fairfax City + Falls Church' \
                                           & State != 'Manassas + Manassas Park' \
                                           & State != 'Buena Vista + Lexington'")
df["State"].unique()
```
