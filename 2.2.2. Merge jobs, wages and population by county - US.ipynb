```python
#open data
dfe = pd.read_csv("data/jobs-by-county-US-refined.csv")
dfw = pd.read_csv("data/income-by-county-US-refined.csv")
dfp = pd.read_csv("data/population-by-county-US-refined.csv")
```

```python
## concat
df = pd.concat([dfe,dfw,dfp], axis = 0).copy()
df
```

```python
df.info()
```

```python
## turn a column into float
df['Value'] = pd.to_numeric(df["Value"], errors = "coerce")
```

```python
df
```

```python
df = df.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)
df = df.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)
```

```python
#Export to CSV
df.to_csv("data/jobs-wages-population-us-merged-by-county.csv", index = False, encoding = "UTF8")
```
