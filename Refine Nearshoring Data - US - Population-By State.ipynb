```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#open manufacturing workers data
df = pd.read_csv("data/population-by-state-US-raw.csv", skiprows=3)
```

```python
#show data
df
```

```python
#drop final rows
df = df.drop([52,53,54,55,56])
```

```python
df
```

```python
#untidy data
df_melted = pd.melt(df, id_vars=["GeoFips","GeoName"]).copy()
df_melted
```

```python
rows = df_melted.shape[0]
rows
```

```python
#arrange columns
df_melted = df_melted.drop("GeoFips", axis =1)
df_melted.rename(columns = {"GeoName":"State","variable": "Year","value": "Population" }, inplace = True)
df_melted
```

```python
#flag border states
border = ["California", "Arizona", "New Mexico", "Texas", "Tamaulipas", "Sonora"]
f = len(border)
df_melted["Border"]="NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_melted["State"][k] == border[l]:
            p = p +1
    if p > 0:
        df_melted["Border"][k] = "Yes"
    else:
        df_melted["Border"][k] = "No"
```

```python
df_melted.groupby("Border")["Border"].agg("count")
```

```python
#Export to CSV
df_melted.to_csv("data/population-by-state-US-refined.csv", index = False, encoding = "UTF8")
```
