```python
#open data
dff = pd.read_csv("data/factories-mexico-refined.csv")
df = pd.read_csv("data/manufacturing-workers-mexico-refined.csv")
dfw = pd.read_csv("data/manufacturing-wages-mexico-refined.csv")
```

```python
dffs = dff.query("City == 'Total'")
dfs = df.query("City == 'Total'")
```

```python
## display both df side by side
display(dffs)
display(dfs)
display(dfw)
```

```python
#merge workers and factories
dffw = pd.merge(dfs, dffs, left_on = ["Year", "Month","State","Border","City"], right_on = ["Year", "Month","State","Border","City"]).drop("City", axis=1)
dffw
```

```python
#merge workers and factories and wages
dffww = pd.merge(dffw, dfw, left_on = ["Year", "Month","State","Border"], right_on = ["Year", "Month","State","Border"])
dffww
```

```python
#arrange columns
dffww = dffww[["Year","Month", "State","Border", "Manufacturing workers","Factories","Manufacturing wages"]]
dffww
```

```python
#Rename columns
dffww.rename(columns = {"Manufacturing workers": "Jobs", "Manufacturing wages": "Average wage", }, inplace = True)
dffww
```

```python
#Annual data (last point of the year)
```

```python
dffwwa = dffww.query("Month == 12").drop("Month", axis=1)
dffwwa
```

```python
#Export to CSV
dffww.to_csv("data/workers-factories-wages-mexico-monthly.csv", index = False, encoding = "UTF8")
dffwwa.to_csv("data/workers-factories-wages-mexico-annual.csv", index = False, encoding = "UTF8")
```
