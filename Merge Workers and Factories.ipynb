```python
#open data
dff = pd.read_csv("data/factories-mexico-refined.csv")
df = pd.read_csv("data/manufacturing-workers-mexico-refined.csv")
```

```python
## display both df side by side
display(dff)
display(df)
```

```python
## merge
dfm = pd.merge(df, dff, left_on = ["Year", "Month","State","City","Border"], right_on = ["Year", "Month","State","City","Border"])
dfm
```

```python
#arrange columns
dfm = dfm[["Year","Month", "State","City", "Border", "Manufacturing workers","Factories"]]
dfm
```

```python
#Annual data (last point of the year)
```

```python
dfa = dfm.query("Month == 12").drop("Month", axis=1)
```

```python
#Export to CSV
dfm.to_csv("data/workers-factories-mexico-monthly.csv", index = False, encoding = "UTF8")
dfa.to_csv("data/workers-factories-mexico-annual.csv", index = False, encoding = "UTF8")
```
