```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#open manufacturing workers data
df1 = pd.read_csv("data/wages-by-count-US-2005-2009-raw.csv", skiprows=3)
df2 = pd.read_csv("data/wages-by-count-US-2010-2015-raw.csv", skiprows=3)
df3 = pd.read_csv("data/wages-by-count-US-2016-2021-raw.csv", skiprows=3)
```

```python
#show data
df1.tail(20)
```

```python
#drop final rows
df1 = df1.drop([3118,3119,3120,3121,3122,3123,3124,3125,3126,3127,3128,3129,3130])
df2 = df2.drop([3118,3119,3120,3121,3122,3123,3124,3125,3126,3127,3128,3129,3130])
df3 = df3.drop([3118,3119,3120,3121,3122,3123,3124,3125,3126,3127,3128,3129,3130])
df3
```

```python
df = pd.concat([df1,df2,df3], axis = 0)
df
```

```python
#untidy data
df_melted = pd.melt(df, id_vars=["GeoFips","GeoName"]).copy()
df_melted
```

```python
#split columns
df_melted["County"] = df_melted["GeoName"].str.split(',', expand=True)[0]
df_melted["State"] = df_melted["GeoName"].str.split(',', expand=True)[1]
```

```python
df_melted
```

```python
#arrange columns
df_melted = df_melted.drop(["GeoFips","GeoName"], axis =1)
df_melted.rename(columns = {"variable": "Year","value": "Population" }, inplace = True)
df_melted = df_melted[["Year", "County","State", "Population"]]
df_melted
```

```python
rows = df_melted.shape[0]
rows
```

```python
df_melted = df_melted.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)
df_melted = df_melted.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)
```

```python
df_melted
```

```python
df_melted["State"].unique()
```

```python
#remove characters
df_melted= df_melted.replace("AK*","AK")
df_melted= df_melted.replace("AZ*","AZ")
df_melted= df_melted.replace("HI*","HI")
df_melted= df_melted.replace("NM*","NM")
df_melted= df_melted.replace("SD*","SD")
df_melted= df_melted.replace("CO*","CO")
df_melted= df_melted.replace("VA*","VA")
df_melted= df_melted.replace("WI*","WI")

df_melted["State"].unique()
```

```python
for i in states:
    df_melted = df_melted.replace([i],states[i])
```

```python
df_melted
```

```python
#flag border states
border = ["California", "Arizona", "New Mexico", "Texas"]
f = len(border)
df_melted["Border State"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_melted["State"][k] == border[l]:
            p = p +1
    if p > 0:
        df_melted["Border State"][k] = "Yes"
    else:
        df_melted["Border State"][k] = "No"
```

```python
#flag border counties
border_counties = ['San Diego', 'Imperial', 'Yuma', 'Pima', 'Santa Cruz', 'Cochise', 'Hidalgo', 'Luna' , 'DoÃ±a Ana','El Paso','Hudspeth' ,'Jeff Davis','Presidio', 'Brewster', 'Terrell' ,'Val Verde', 'Kinney' ,'Maverick', 'Webb' ,'Zapata', 'Starr' ,'Hidalgo', 'Cameron']
f = len(border_counties)
df_melted["Border County"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_melted["County"][k] == border_counties[l]:
            p = p +1
    if p > 0:
        df_melted["Border County"][k] = "Yes"
    else:
        df_melted["Border County"][k] = "No"
```

```python
#Remove non states
df_melted = df_melted.query("State != 'Staunton + Waynesboro' \
                                           & State != 'Colonial Heights + Petersburg' \
                                           & State != 'Fairfax City + Falls Church' \
                                           & State != 'Manassas + Manassas Park' \
                                           & State != 'Buena Vista + Lexington'")
df_melted["State"].unique()
```
