```python
#untidy data
df_melted = pd.melt(df_concat, id_vars=['GeoName','Description'], var_name='Year', value_name='Value')
df_melted.reset_index()
```

```python
#split columns
df_melted["County"] = df_melted["GeoName"].str.split(',', expand=True)[0]
df_melted["State"] = df_melted["GeoName"].str.split(',', expand=True)[1]
```

```python
#arrange columns
df_melted = df_melted.drop(["GeoName"], axis =1)
df_melted = df_melted[["Year", "County","State","Description", "Value"]]
df_melted
```

```python
rows = df_melted.shape[0]
rows
```

```python
#flag border states
border = ["California", "Arizona", "New Mexico", "Texas"]
f = len(border)
df_melted["Border State"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_melted["State"][k] == border[l]:
            p = p +1
    if p > 0:
        df_melted["Border State"][k] = "Yes"
    else:
        df_melted["Border State"][k] = "No"
```
