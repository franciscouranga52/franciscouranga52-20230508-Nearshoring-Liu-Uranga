```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#open manufacturing workers data
df = pd.read_csv("data/total-jobs-by-state-mexico.csv", skiprows=4)
```

```python
#show data
df
```

```python
#drop final rows
df = df.drop([68,69])
```

```python
df
```

```python
#untidy data
df_melted = pd.melt(df, id_vars=["Periodos"], var_name ="City", value_name ="Jobs").copy()
df_melted
```

```python
rows = df_melted.shape[0]
rows
```

```python
df_melted["City"][220]
```

```python
#clean data (parentesis)
for i in range(rows):
    df_melted["City"][i]= re.sub(r'\([^)]*\)', '', df_melted["City"][i])
```

```python
#clean data (other text)
df_clean= df_melted.replace("Ocupaci�n, empleo y remuneraciones > Poblaci�n ocupada, subocupada y desocupada  > Valores absolutos > Por ciudad > Total > ","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Desocupada /a /f1'","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n de 15 a�os y m�s /a /f1","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n subocupada /a /f1","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n subocupada /f1 /b","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Total /a /f1","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Ocupada /a /f1","",regex=True)
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Desocupada /a /f1","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n de 15 a�os y m�s /f1 /b","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Ocupada /f1 /b","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Desocupada /f1 /b","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Total /f1 /b","",regex=True) 
df_clean= df_clean.replace(" > Poblaci�n econ�micamente activa  > Subocupada /f1 /b","",regex=True) 
df_clean= df_clean.replace(">",",",regex=True)
df_clean
```

```python
df_clean["City"].unique()
```

```python
#clean other characters
df_clean = df_clean.replace("Canc�n","Cancun",regex=True)
df_clean = df_clean.replace("Ju�rez","Juarez",regex=True)
df_clean = df_clean.replace("Culiac�n","Culiacan",regex=True)
df_clean = df_clean.replace("M�xico","Mexico",regex=True)
df_clean = df_clean.replace("Le�n","Leon",regex=True)
df_clean = df_clean.replace("M�rida","Merida",regex=True)
df_clean = df_clean.replace("Quer�taro","Queretaro",regex=True)
df_clean = df_clean.replace("Potos�","Potosi",regex=True)
df_clean = df_clean.replace("Yucat�n","Yucatan",regex=True)
df_clean = df_clean.replace("Torre�n","Torreon",regex=True)
df_clean = df_clean.replace("Guti�rrez","Gutierrez",regex=True)
df_clean["City"].unique()
```

```python
#split columns
df_clean["Quarter"] = df_clean["Periodos"].str.split('/', expand=True)[1]
df_clean["Year"] = df_clean["Periodos"].str.split('/', expand=True)[0]
df_clean.drop(["Periodos"], axis=1)
```

```python
#arrange columns
df_clean = df_clean[["Year","Quarter","City", "Jobs"]]
df_clean
```

```python
df_clean["City"].unique()
```

```python
#remove blank spaces at the end
for a in range(rows):
    df_clean["City"][a] = df_clean["City"][a].rstrip().lstrip()
```

```python
df_clean.sample(20)
```

```python
df_clean["City"].unique()
```

```python
#flag cities
border = ["Ciudad Juarez", "Hermosillo", "Chihuahua", "Leon", "Mexicali", "Monterrey", "Reynosa", "Saltillo","Tampico","Tijuana", "Torreon"]
f = len(border)
df_clean["Border"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_clean["City"][k] == border[l]:
            p = p +1
    if p > 0:
        df_clean["Border"][k] = "Yes"
    else:
        df_clean["Border"][k] = "No"
```
