```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#open manufacturing workers data
df = pd.read_csv("data/GDP by industry -Mexico.csv", skiprows=4)
```

```python
#show data
df
```

```python
#drop final rows
df = df.drop([42,43])
```

```python
#untidy data
df_melted = pd.melt(df, id_vars=["Periodos"], var_name ="State", value_name ="GDP").copy()
df_melted
```

```python
rows = df_melted.shape[0]
rows
```

```python
df_melted["State"][220]
```

```python
#clean data (parentesis)
for i in range(rows):
    df_melted["State"][i]= re.sub(r'\([^)]*\)', '', df_melted["State"][i])
```

```python
df_melted["State"][220]
```

```python
#clean data (other text)
df_clean= df_melted.replace("Cuentas nacionales > Producto interno bruto por entidad federativa, base 2013 > Por actividad econ�mica y entidad federativa > Valores a precios constantes de 2013 > Actividades secundarias > Total actividades secundarias > ","",regex=True)
df_clean= df_clean.replace(" /f1","",regex=True)
df_clean= df_clean.replace(" /f1",",",regex=True)
df_clean
```

```python
df_clean["State"].unique()
```

```python
#clean other characters
df_clean = df_clean.replace("M�xico","Mexico",regex=True)
df_clean = df_clean.replace("Michoac�n","Michoacan",regex=True)
df_clean = df_clean.replace("Quer�taro","Queretaro",regex=True)
df_clean = df_clean.replace("Potos�","Potosi",regex=True)
df_clean = df_clean.replace("Le�n","Leon",regex=True)
df_clean = df_clean.replace("Yucat�n","Yucatan",regex=True)
```

```python
df_clean.sample(20)
```

```python
df_clean["State"].unique()
```

```python
#remove blank spaces at the end
for a in range(rows):
    df_clean["State"][a] = df_clean["State"][a].rstrip().lstrip()
```

```python
df_clean.sample(20)
```

```python
#flag border states and cities
border = ["Baja California", "Coahuila de Zaragoza", "Chihuahua", "Nuevo Leon", "Tamaulipas", "Sonora"]
f = len(border)
df_clean["Border"] = "NA"
for k in range(rows):
    p = 0
    for l in range(f):
        if df_clean["State"][k] == border[l]:
            p = p +1
    if p > 0:
        df_clean["Border"][k] = "Yes"
    else:
        df_clean["Border"][k] = "No"
```

```python
df_clean.sample(20)
```

```python
df_clean.groupby("Border")["Border"].agg("count")
```

```python
#Export to CSV
df_clean.to_csv("data/GDP-mexico-refined.csv", index = False, encoding = "UTF8")
```
