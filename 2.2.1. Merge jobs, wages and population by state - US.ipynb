```python
#open data
dfe = pd.read_csv("data/employment-by-state-US-refined.csv")
dfw = pd.read_csv("data/wages-by-state-US-refined.csv")
dfp = pd.read_csv("data/population-by-state-US-refined.csv")
```

```python
dfe = dfe.replace("Private nonfarm employment", "Total Jobs",regex=True)
dfe = dfe.replace("Construction", "Construction Jobs",regex=True)
dfe = dfe.replace("Manufacturing", "Manufacturing Jobs",regex=True)
dfe = dfe.replace("Transportation and warehousing", "Transportation Jobs",regex=True)
dfe.rename(columns = {"Jobs":"Value"}, inplace = True)
```

```python
dfw = dfw.replace("Private nonfarm wages and salaries", "Total Wages",regex=True)
dfw = dfw.replace("Construction", "Construction Wages",regex=True)
dfw = dfw.replace("Manufacturing", "Manufacturing Wages",regex=True)
dfw = dfw.replace("Transportation and warehousing", "Transportation Wages",regex=True)
dfw.rename(columns = {"Wages":"Value"}, inplace = True)
```

```python
dfp["Description"] = "Population"
dfp.rename(columns = {"Population":"Value"}, inplace = True)
```

```python
## display df side by side
display(dfe)
display(dfw)
display(dfp)
```

```python
## concat
df = pd.concat([dfe,dfw,dfp], axis = 0).copy()
df
```

```python
df.info()
```

```python
## turn a column into float
df['Value'] = pd.to_numeric(df["Value"], errors = "coerce")
```

```python
df
```

```python
df = df.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)
df = df.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)
```
